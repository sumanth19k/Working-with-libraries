## axis, legends

```python
day = [1,2,3,4,5,6,7]
max_t = [41,43,42,44,43,41,45]
min_t = [22,23,25,21,27,30,23]
```

```python
plt.plot(day, max_t, label='Max')
plt.plot(day, min_t, label='Min')

plt.legend()
```

```python
plt.xlabel("days", fontsize="large")
plt.ylabel("temperature", fontsize="large")
plt.title("Weather report")

plt.plot(day, max_t, label='Max')
plt.plot(day, min_t, label='Min')

plt.legend(loc="best", shadow=True, fontsize="medium")
```

```python
plt.xlabel("days", fontsize="large")
plt.ylabel("temperature", fontsize="large")
plt.title("Weather report")

plt.plot(day, max_t, label='Max')
plt.plot(day, min_t, label='Min')

plt.legend(loc="best", shadow=True, fontsize="medium")
plt.grid()
```

## barchart

```python
company = ['Microsoft', 'Google', 'Tesla', 'Amazon']
revenue = [90, 136, 89, 27]
profit = [50, 70, 120, 5]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue")
plt.title("US tech stocks")
plt.bar(xpos, revenue, label="stocks")

plt.legend(shadow=True)
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue")
plt.title("US tech stocks")
plt.bar(xpos-0.2, revenue,width=0.4, label="stocks")
plt.bar(xpos+0.2, profit,width=0.4, label="profits")
plt.legend(shadow=True)
```

```python
plt.yticks(xpos, company)
plt.xlabel("comaprison")
plt.title("US tech stocks")

plt.barh(xpos-0.2, revenue,height = 0.4, label="stocks")
plt.barh(xpos+0.2, profit,height=0.4, label="profits")
plt.legend(shadow=True)
```
